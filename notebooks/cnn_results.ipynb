# Gunshot CNN Results

Display-only notebook for showcasing gunshot CNN training run results.
Figures are pre-generated by `elp_gunshot.evaluate_cnn`.

This notebook is pinned to the current best run:
`model3_nomask_bs64_lr5e-05_e70_20260318_180802`

To regenerate figures for a run:
```bash
python -m elp_gunshot.evaluate_cnn --run_dir runs/<run_name>
```

````python
import json
from pathlib import Path

import pandas as pd
from IPython.display import Image, display

BEST_RUN = "model3_nomask_bs64_lr5e-05_e70_20260318_180802"
RUN_DIR = Path("../runs") / BEST_RUN
FIGURES_DIR = Path("../results") / BEST_RUN
````

## Test Metrics Summary

````python
metrics_path = RUN_DIR / "test_metrics.json"
if not metrics_path.exists():
    raise FileNotFoundError(f"Metrics file not found: {metrics_path}")

with open(metrics_path) as f:
    metrics = json.load(f)

rows = [(k, v) for k, v in metrics.items() if k != "confusion_matrix"]
pd.DataFrame(rows, columns=["Metric", "Value"]).set_index("Metric")
````
