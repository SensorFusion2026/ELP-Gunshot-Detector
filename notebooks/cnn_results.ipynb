# Gunshot CNN Results

Display-only notebook for showcasing gunshot CNN training run results.
Figures are pre-generated by `elp_gunshot.evaluate_cnn`.

To regenerate figures for a run:
```bash
python -m elp_gunshot.evaluate_cnn --run_dir runs/<run_name>
```

````python
import json
from pathlib import Path

import pandas as pd
from IPython.display import Image, display

FIGURES_DIR = Path("../results/figures")

# Point this at a specific run directory to pin results.
RUN_DIR = Path("../runs")
````

## Test Metrics Summary

````python
if not (RUN_DIR / "test_metrics.json").exists():
    runs = sorted(RUN_DIR.glob("*/test_metrics.json"), key=lambda p: p.stat().st_mtime)
    if not runs:
        raise FileNotFoundError("No completed runs found. Run training first.")
    metrics_path = runs[-1]
    print(f"Using run: {metrics_path.parent.name}")
else:
    metrics_path = RUN_DIR / "test_metrics.json"

with open(metrics_path) as f:
    metrics = json.load(f)

rows = [(k, v) for k, v in metrics.items() if k != "confusion_matrix"]
pd.DataFrame(rows, columns=["Metric", "Value"]).set_index("Metric")
````
